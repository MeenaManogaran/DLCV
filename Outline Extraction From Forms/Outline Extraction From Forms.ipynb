```python
import argparse
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread(r"C:\Users\Meena\OneDrive - Alliance\Desktop\DLCV\data\form\image.png")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot(1, 2, 1)
plt.imshow(image);
plt.title('Original')

kernel = np.ones((0,4),np.uint8)
dilation = cv2.dilate(~gray.copy(),kernel,iterations = 1)
h_kernel = np.ones((90,1),np.uint8)
h_lines = cv2.morphologyEx(dilation, cv2.MORPH_OPEN, h_kernel)
v_kernel = np.ones((1,80),np.uint8)
v_lines = cv2.morphologyEx(dilation, cv2.MORPH_OPEN, v_kernel)
line_img = h_lines + v_lines

plt.subplot(1, 2, 2)
plt.imshow(line_img);
plt.title('Extracted Line Image');
```
