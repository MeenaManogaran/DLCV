```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
import glob
images = []
for img in glob.glob(r'C:\Users\Meena\OneDrive - Alliance\Desktop\DLCV\data\Dhoni\*.jpg'):
    n= cv2.imread(img)
    mask = np.zeros(n.shape[:2],np.uint8)
    
    bgdModel = np.zeros((1,65),np.float64)
    fgdModel = np.zeros((1,65),np.float64)
    rect = (13,4,230,300)
    cv2.grabCut(n,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
    mask2 = np.where((mask==0)|(mask==2),0,1).astype('uint8')
    imgs= n*mask2[:,:,np.newaxis]
    plt.imshow(imgs),plt.colorbar(),plt.show()

    images.append(n)
    #print (img)
```
